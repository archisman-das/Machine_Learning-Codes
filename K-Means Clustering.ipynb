```python
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("income.csv")
df.head()
```

```python
plt.scatter(df['Age'],df['Income($)'])
plt.show()
```

```python
km = KMeans(n_clusters=3)
km
```

```python
y_predicted=km.fit_predict(df[['Age','Income($)']])
y_predicted
```

```python
df['cluster'] = y_predicted
df.head()
```

```python
df1= df[df.cluster==0]
df2= df[df.cluster==1]
df3= df[df.cluster==2]

plt.scatter(df1.Age,df1['Income($)'],color = 'green')
plt.scatter(df2.Age,df2['Income($)'],color = 'red')
plt.scatter(df3.Age,df3['Income($)'],color = 'black')

plt.xlabel('Age')
plt.ylabel('Income($)')
plt.legend()
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(df[['Income($)']])  # Fitting the scaler on the DataFrame
df['Income($)'] = scaler.transform(df[['Income($)']])  # Use double brackets to pass a DataFrame


scaler.fit(df[['Age']])  # Use double brackets to pass a DataFrame
df['Age'] = scaler.transform(df[['Age']])  # Use double brackets to pass a DataFrame
df
```

```python
km = KMeans(n_clusters=3)
y_predicted = km.fit_predict(df[['Age','Income($)']])
y_predicted
```

```python
df['cluter'] = y_predicted
df.drop('cluter',axis='columns',inplace = True)
df
```

```python
# Ensure that the DataFrame 'df' has a column named 'cluster'.
# Replace 'cluster' with the actual name of the column if it's different.

# Example: If the column is named 'group', use df.group instead of df.cluster
df1 = df[df['cluster'] == 0]  # Check if 'cluster' is the correct column name
df2 = df[df['cluster'] == 1]
df3 = df[df['cluster'] == 2]

plt.scatter(df1.Age, df1['Income($)'], color='green')
plt.scatter(df2.Age, df2['Income($)'], color='red')
plt.scatter(df3.Age, df3['Income($)'], color='black')

plt.legend()
plt.show()
```
